```python
# очищаем от дубликатов
df = df[~df.duplicated()] # очистка от дубликатов
display(df)
```

```python
X = df.drop(['target'], axis = 1)
y = df.target
```

```python
axis_features = ['age','sex','trestbps','chol','fbs', 'restecg','thalach','exang','oldpeak','slope', 'ca', 'thal', 'cp'] # числовые

imputer_for_nums = Pipeline(steps=[
    ('imputer', KNNImputer(n_neighbors = 5, missing_values=np.nan)), # на всякий случай
    ('scaler', MinMaxScaler())
])

CT = ColumnTransformer([
    
        ("imputer_for_nums", imputer_for_nums, axis_features), # здесь идет преобразование числовых
             ], 
    
        remainder = 'passthrough')

display(CT)

res_ct = CT.fit_transform(X)
```

```python
# Блок предсказания с использованием тестового набора
```

```python
df_test = pd.read_csv(path_test) 
```

```python
df_test
```

```python
pipe = Pipeline([ ('preprocessing', CT) ,
                 ('svc', LogisticRegression(C = 5, penalty = 'l2', solver = 'newton-cg' ))])
pipe.fit(X,y)
print(roc_auc_score(y, pipe.predict_proba(X)[:, 1]))
# print(roc_auc_score(y_test, pipe.predict_proba(X_test)[:, 1]))

# Есть небольшое переобучение, но в целом норм
```

```python
y_predict = pipe.predict(df_test)
```
